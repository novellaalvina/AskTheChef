# Intent Classification

```python
from transformers import pipeline
def intent_classification_with_model(user_input):
  classifier = pipeline("zero-shot-classification", model="facebook/bart-large-mnli", device=-1)
  input_text = user_input
  labels = ["recipe_request", "general_conversation"]
  result = classifier(input_text, labels, multi_label=False)
  return result["labels"][0]

# intent_classification_with_model("I am planning to host a thanksgiving party tonight. I have a turkey but I don't know what to do with it. Do you have suggestions on how to cook the turkey? Also, can you suggest some fun thanksgiving activities for the party?")
# intent_classification_with_model("hi, how are you?")
```

# dialogue manager

```python
def classify_intent(user_input):
    # If user mentions keywords related to cooking, assume they want a recipe
    recipe_keywords = ["recipe", "cook", "how to make", "ingredients", "method", "how do I cook"]
    if any(keyword in user_input.lower() for keyword in recipe_keywords):
        return "recipe_request"
    return "general_conversation"
```

```python
def post_process_recipe(response):
    sentences = response.split('.')
    unique_sentences = list(dict.fromkeys(sentences))  # Remove duplicates
    return '. '.join(unique_sentences) + '.'

# response = post_process_recipe(recipe_tokenizer.decode(output[0], skip_special_tokens=True))
```

```python
from rapidfuzz import process, fuzz

def recipe_request_processing(user_input):
  # Example: List of dish names
  dish_names = list(train_dataset['name'])

  # User input
  user_input = "noodle soup with pork chop"

  # Find the best match with a similarity threshold
  best_match, similarity, index = process.extractOne(query=user_input, choices=dish_names, scorer=fuzz.WRatio)
  if similarity > 80:  # Threshold for similarity
      print(f"Match found: {best_match} with similarity {similarity}%")
      desc = train_dataset[train_dataset['name'] == best_match]['description'].values[0]
      ingredients = train_dataset[train_dataset['name'] == best_match]['ingredients'].values[0]
      steps = train_dataset[train_dataset['name'] == best_match]['steps'].values[0]
      ingredients = ingredients.split(',')
      steps = steps.split('.')

      print(f"Description: {desc}")
      print("\n")
      print(f"Ingredients:\n")
      for ingredient in ingredients:
        print(ingredient)
      print("\n")
      print(f"Steps:\n")
      for step in steps:
        print(step)
      return True
  else:
      return False
```

```python
def post_process_response(response):
    sentences = response.split('. ')
    seen_sentences = set()
    processed_sentences = []
    for sentence in sentences:
        if sentence not in seen_sentences:
            processed_sentences.append(sentence)
            seen_sentences.add(sentence)
    return '. '.join(processed_sentences).strip()
```

```python
def dialogue_manager(user_input):
    # Classify the user's intent
    intent = classify_intent(user_input)

    if intent == "recipe_request":

        input_dish = recipe_request_processing(user_input)
        if input_dish is False:
          # Use RecipeNLG model to generate a response
          inputs = recipe_tokenizer.encode(user_input, return_tensors="pt")
          output = recipe_model.generate(inputs, max_length=150, num_return_sequences=1, pad_token_id=recipe_tokenizer.eos_token_id)
          response = recipe_tokenizer.decode(output[0], skip_special_tokens=True)

    elif intent == "general_conversation":
        # Use Llama model for general chat (set max_token=0 to remove limit on max token)
        response = llm(user_input, max_tokens=0)

        # display the response
        response = response["choices"][0]["text"]

    else:
        response = "I'm not sure how to respond to that. Could you clarify?"

    return response
```

```python
MAX_HISTORY_LENGTH = 3  # Limit to last 3 turns

conversation_history = []

def dialogue_manager_with_limited_history(user_input, conv_model):
    try:
        # Update conversation history
        conversation_history.append(f"User: {user_input}")

        # Classify intent
        intent = intent_classification_with_model(user_input)
        print("intent: ", intent)

        # Limit conversation history to keep context concise
        if len(conversation_history) > MAX_HISTORY_LENGTH * 2:  # Each turn is user+bot
            conversation_history.pop(0)  # Keep history length reasonable

        if intent == "recipe_request":

            # if the user input the dish name
            input_dish = recipe_request_processing(user_input)
            if input_dish:
              response = ""
              return response

            # if the user input the list of ingredients
            else:
              # Modify the prompt for clarity and ensure it's concise
              prompt = user_input + " Please provide a concise recipe response."

              # Generate a recipe response
              inputs = recipe_tokenizer.encode(prompt, return_tensors="pt")

              # Catch errors related to tokenization or input length
              if inputs is None or inputs.size()[1] > 512:
                  return "Sorry, the input is too long for me to process. Could you please rephrase or provide fewer details?"

              output = recipe_model.generate(
                      inputs,
                      max_length=300,
                      num_return_sequences=1,
                      pad_token_id=recipe_tokenizer.eos_token_id,
                      temperature=0.7,  # Less randomness for consistency
                      top_k=50,
                      top_p=0.9,
                      repetition_penalty=1.2,
                      do_sample=False,  # Disable sampling
                      num_beams=3,  # Enable beam search for more coherent results
                      early_stopping=True  # Stops when a coherent point is reached
                  )

              # Check for valid output
              if output is None:
                  return "I couldn't generate a recipe response. Could you try asking again?"

              response = recipe_tokenizer.decode(output[0], skip_special_tokens=True)
              response = post_process_recipe(response)

        elif intent == "general_conversation":

          if conv_model == "llama":

            # Use conversation history to maintain context
            context_input = "\n".join(conversation_history[-MAX_HISTORY_LENGTH:]) + f"\nAI:"
            response = llm(context_input, max_tokens=150)

            # Extract response from Llama's output and handle None response
            if response is None or not isinstance(response, dict) or 'choices' not in response:
                return "I'm having trouble generating a response right now. Could you please rephrase your question?"

            response_text = response['choices'][0]['text'] if 'choices' in response else response
            conversation_history.append(f"AI: {response_text}")
            return response_text

          elif conv_model == "distilgpt-2":
            prompt = """You are a helpful assistant. Provide thoughtful and relevant responses to users.
              """ + "\n".join(conversation_history[-MAX_HISTORY_LENGTH:]) + f"\nUser: {user_input}\nAssistant:"

            # Tokenize input with padding and attention mask
            inputs = dgpt_tokenizer(prompt, return_tensors="pt", padding=True)

            # Generate response with improved constraints
            output = dgpt2.generate(
                inputs.input_ids,
                attention_mask=inputs.attention_mask,
                max_length=260,
                temperature=0.7,
                top_k=50,
                top_p=0.9,
                repetition_penalty=2.0,  # Penalize repeating content
                no_repeat_ngram_size=3,  # Prevent repeated 3-grams
                do_sample=True,
                num_return_sequences=1
            )
            response = dgpt_tokenizer.decode(output[0], skip_special_tokens=True)
            response = post_process_response(response)

            return response

          elif conv_model == "gpt-2":
            # Instruction and user input
            instruction = "You are a helpful assistant. Please provide thoughtful and concise responses."

            # Limit conversation history (if available) and format prompt
            context = ""
            if len(conversation_history) > 0:
                context = "\n".join(conversation_history[-MAX_HISTORY_LENGTH:])

            # Construct prompt with instruction, recent history, and user input
            prompt = f"{instruction}\n{context}\nUser: {user_input}\nAssistant:"

            # Tokenize input with padding and attention mask
            inputs = gpt_tokenizer(prompt, return_tensors="pt", padding=True, truncation=True)

            # Generate response with GPT-2
            output = gpt2.generate(
                inputs.input_ids,
                attention_mask=inputs.attention_mask,
                max_length=200,
                temperature=0.5,
                top_k=50,
                top_p=0.9,
                repetition_penalty=3.0,  # Penalize repetition further
                no_repeat_ngram_size=3,
                do_sample=True,
                num_return_sequences=1,
                num_beams=3,
                eos_token_id=gpt_tokenizer.eos_token_id
            )

            # Decode and post-process response
            response = gpt_tokenizer.decode(output[0], skip_special_tokens=True)
            response = post_process_response(response)

            # Add the new response to the conversation history, ensuring history remains concise
            conversation_history.append(f"User: {user_input}")
            conversation_history.append(f"Assistant: {response}")

            return response
        else:
            return "I'm not sure how to respond to that. Could you clarify?"

    except Exception as e:
        # Handle general exceptions and provide a safe fallback message
        return f"An error occurred: {str(e)}. Please try again."

    return response
```

```python
# Simple chat loop
import timeit
print("Chatbot is ready! Type 'exit' to end the chat.")
conversation_history = []

conv_model = input("Which conversational model do you want to use? Llama or DistilGPT-2 or GPT-2? ").lower()
print(type(conv_model))
while conv_model != "llama" and conv_model != "distilgpt-2" and conv_model != "gpt-2":
  print("please only choose either Llama or DistilGPT-2 or GPT-2.")
  conv_model = input("Which conversational model do you want to use? Llama or DistilGPT-2 or GPT-2? ").lower()

start = timeit.default_timer()
print(f"start time: {start}")
while True:
    user_input = input("You: ")
    input_time = timeit.default_timer()
    print(f"input time: {input_time}")
    if user_input.lower() == "exit":
        break
    response = dialogue_manager_with_limited_history(user_input, conv_model)
    print(f"Chatbot: {response['choices'][0]['text'] if 'choices' in response else response}")
    response_time = timeit.default_timer()
    print(f"response time: {response_time}")
end = timeit.default_timer()
print(f"end time: {end}")
print(f"This conversation takes {end-start} seconds.")
```

```python
import nltk
from rouge_score import rouge_scorer
from sklearn.metrics import classification_report
from sentence_transformers import SentenceTransformer, util
from transformers import AutoTokenizer, AutoModelForCausalLM
import torch
```

```python
# embedding model for similarity metrics
embedding_model = SentenceTransformer('sentence-transformers/all-MiniLM-L6-v2')
```

```python
# BLEU Score
def compute_bleu(reference, hypothesis):
    reference = [nltk.word_tokenize(reference.lower())]
    hypothesis = nltk.word_tokenize(hypothesis.lower())
    bleu_score = nltk.translate.bleu_score.sentence_bleu(reference, hypothesis)
    return bleu_score

# ROUGE Score
def compute_rouge(reference, hypothesis):
    scorer = rouge_scorer.RougeScorer(['rouge1', 'rouge2', 'rougeL'], use_stemmer=True)
    scores = scorer.score(reference, hypothesis)
    return scores

# Cosine Similarity
def compute_similarity(reference, hypothesis):
    ref_embedding = embedding_model.encode(reference, convert_to_tensor=True)
    hyp_embedding = embedding_model.encode(hypothesis, convert_to_tensor=True)
    similarity = util.pytorch_cos_sim(ref_embedding, hyp_embedding).item()
    return similarity
```

```python
# Recipe Generation
def generate_recipe(prompt):
    inputs = recipe_tokenizer(prompt, return_tensors="pt")
    output = recipe_model.generate(
        inputs.input_ids,
        max_length=150,
        temperature=0.7,
        top_k=50,
        top_p=0.9,
        repetition_penalty=1.2,
        do_sample=True,
        num_return_sequences=1
    )
    response = recipe_tokenizer.decode(output[0], skip_special_tokens=True)
    return response
```

```python
# Conversational Model Comparison
def evaluate_models(input_text, models):
    results = {}
    for model_name, model_info in models.items():
        if model_name == "Llama":
            # Generate response using Llama
            response = llm(input_text)["choices"][0]["text"]
        else:
            # Generate response using GPT models
            tokenizer, model = model_info["tokenizer"], model_info["model"]
            inputs = tokenizer(input_text, return_tensors="pt", padding=True, truncation=True)
            output = model.generate(
                inputs.input_ids,
                attention_mask=inputs.attention_mask,
                max_length=150,
                temperature=0.7,
                top_k=50,
                top_p=0.9,
                repetition_penalty=1.2,
                do_sample=True
            )
            response = tokenizer.decode(output[0], skip_special_tokens=True)
        results[model_name] = response.strip()
    return results
```

```python
# Initialize models for conversational evaluation
models = {
    "DistilGPT-2": {
        "tokenizer":dgpt_tokenizer,
        "model": dgpt2
    },
    "GPT-2": {
        "tokenizer": gpt_tokenizer,
        "model": gpt2
    },
    "Llama": llm
}
```

```python
# Example evaluation
user_input = "I want a recipe for a scrambled egg."
expected_output = """Scrambled Eggs Recipe:

Ingredients:
- 2 large eggs
- 1 tbsp milk (optional)
- 1 tbsp butter or oil
- Salt and pepper to taste

Instructions:
1. Crack eggs into a bowl. Add milk (for creaminess) and whisk until well blended.
2. Heat a non-stick pan over medium heat and melt butter or add oil.
3. Pour the egg mixture into the pan.
4. Let it sit for a few seconds, then gently stir with a spatula, pushing from the edges to the center.
5. Cook until eggs are softly set but slightly runny for creamy texture.
6. Remove from heat (eggs will continue cooking).
7. Season with salt and pepper. Serve immediately.
"""

# Generate recipe
generated_recipe = generate_recipe(user_input)
print(f"Generated Recipe: {generated_recipe}")

# Compute Evaluation Metrics
bleu = compute_bleu(expected_output, generated_recipe)
rouge = compute_rouge(expected_output, generated_recipe)
similarity = compute_similarity(expected_output, generated_recipe)

print(f"BLEU Score: {bleu}\n")
print(f"ROUGE Score: {rouge}\n")
print(f"Semantic Similarity: {similarity}\n")

# Evaluate conversational models
conversation_input = "Hi, how are you?"
conversation_output = df["answer"][0]
model_responses = evaluate_models(conversation_input, models)
print("Model Responses:\n")
for model_name, response in model_responses.items():
    print(f"{model_name}: {response}\n")

    # Compute Evaluation Metrics
    bleu = compute_bleu(conversation_output, response)
    rouge = compute_rouge(conversation_output, response)
    similarity = compute_similarity(conversation_output, response)

    print(f"model name: {model_name}\n")
    print(f"BLEU Score: {bleu}\n")
    print(f"ROUGE Score: {rouge}\n")
    print(f"Semantic Similarity: {similarity}\n")
```

```python
from transformers import AutoTokenizer, AutoModelForCausalLM

# Load DistilGPT-2 or GPT-2
model_name = "distilgpt2"  # Change to "gpt2" if you prefer GPT-2
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name)

# Function to generate conversational responses
def gpt_conversation(user_input):
    try:
        inputs = tokenizer(user_input, return_tensors="pt")
        output = model.generate(
            inputs.input_ids,
            max_length=150,  # Adjust for desired response length
            temperature=0.7,  # Adds some variation
            top_k=50,  # Limits sampling pool
            top_p=0.9,  # Limits to most probable tokens
            repetition_penalty=1.2,  # Penalizes repetitive outputs
            do_sample=True,  # Enable sampling for creative output
            num_return_sequences=1
        )
        response = tokenizer.decode(output[0], skip_special_tokens=True)
        return response
    except Exception as e:
        return f"An error occurred with the GPT response: {str(e)}"

# Example of usage in dialogue manager
user_input = "What are some good ways to cook pasta?"
response = gpt_conversation(user_input)
print(f"Chatbot: {response}")
```
